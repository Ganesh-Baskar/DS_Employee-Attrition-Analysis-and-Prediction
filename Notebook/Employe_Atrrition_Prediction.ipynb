```python
# Encode the 'Attrition' column: Yes → 1, No → 0
df["Attrition"] = df["Attrition"].replace({"Yes": 1, "No": 0})
```

```python
# Create new feature representing estimated tenure excluding time with current manager
df["EstimatedTenure"] = df["Age"] - df["YearsWithCurrManager"]

# Combine tenure and job satisfaction to assess potential engagement
df["EngagementScore"] = df["EstimatedTenure"] * df["JobSatisfaction"]

# Composite feature reflecting age and seniority level
df["SeniorityIndex"] = df["Age"] * df["JobLevel"]
```

```python
# Drop columns that are either constant, identifiers, or show multicollinearity
columns_to_drop = [
    "EmployeeCount",        # Constant value across all rows
    "EmployeeNumber",       # Unique identifier, not useful for modeling
    "JobLevel",             # Highly correlated with Age or Income
    "TotalWorkingYears",    # Likely overlaps with YearsAtCompany or Age
    "YearsAtCompany",       # Potential collinearity with tenure-related features
    "YearsInCurrentRole",   # Might overlap with YearsWithCurrManager
    "Over18",               # Constant categorical value
    "StandardHours"         # All employees typically have the same hours
]

df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
# Separate features and target variable
X = df.drop("Attrition", axis=1)  # Feature set
y = df["Attrition"]               # Target variable: Attrition status
```

```python
# Define feature categories based on data type and semantic meaning

# Continuous and discrete numerical features
numerical_cols = [
    "Age", "DailyRate", "DistanceFromHome", "Education", "TrainingTimesLastYear",
    "EnvironmentSatisfaction", "HourlyRate", "JobInvolvement", "YearsWithCurrManager",
    "MonthlyIncome", "MonthlyRate", "NumCompaniesWorked", "PercentSalaryHike",
    "RelationshipSatisfaction", "StockOptionLevel", "WorkLifeBalance", "YearsAtCompany",
    "Tenure", "Tenure_JobSatisfaction", "Age_JobLevel", "YearsSinceLastPromotion"
]

# Categorical features (non-ordered)
categorical_cols = [
    "BusinessTravel", "Department", "Gender", "JobRole", "OverTime"
]

# Ordinal features (ordered categories)
ordinal_cols = [
    "JobSatisfaction", "PerformanceRating"
]
```

```python
#  Step 1: Scale numerical features using StandardScaler
from sklearn.preprocessing import StandardScaler

# Ensure only existing numerical columns are selected
valid_numerical_cols = [col for col in numerical_cols if col in df.columns]

# Initialize and apply the scaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[valid_numerical_cols])

# Convert scaled data to a DataFrame with appropriate column names
scaled_df = pd.DataFrame(scaled_data, columns=scaler.get_feature_names_out(valid_numerical_cols))

# Display the first few rows of the scaled numerical features
scaled_df.head()
```

```python
# Step 2: One-hot encode categorical features
from sklearn.preprocessing import OneHotEncoder

# Initialize the encoder with options for robust transformation
encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

# Apply encoding to selected categorical columns
encoded_array = encoder.fit_transform(df[categorical_cols])

# Convert the encoded array to a DataFrame with appropriate column names
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(categorical_cols))

# Preview the encoded categorical features
encoded_df.head()
```

```python
# Step 3: Encode ordinal features using OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder

# Initialize and apply the ordinal encoder
ordinal_encoder = OrdinalEncoder()
ordinal_encoded_array = ordinal_encoder.fit_transform(df[ordinal_cols])

# Convert the result to a DataFrame with original column names
ordinal_encoded_df = pd.DataFrame(ordinal_encoded_array, columns=ordinal_cols)

# Preview the encoded ordinal features
ordinal_encoded_df.head()
```

```python
# Step 4: Combine all transformed feature sets into one DataFrame
X_transformed = pd.concat(
    [scaled_df, encoded_df, ordinal_encoded_df],
    axis=1
)

# Preview the final transformed feature set
X_transformed.head()
```

```python
# Step 6: Save final processed dataset with target variable included
final_df = X_transformed.copy()

# Add the target variable back to the transformed dataset
final_df["Attrition"] = df["Attrition"]

# Optionally save to CSV
final_df.to_csv("processed_attrition_data.csv", index=False)

# Preview the final dataset
final_df.head()
```

```python
# Step 7: Remove any rows with missing values
final_df.dropna(inplace=True)
```

```python
# Save the final processed dataset to CSV
final_csv_path = "processed_employee_data.csv"
final_df.to_csv(final_csv_path, index=False)

print(f"\n Successfully saved processed data to: {final_csv_path}")
```
